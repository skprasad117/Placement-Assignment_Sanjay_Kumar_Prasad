```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv("Dataset/statistics_solution_6_data.csv")
df.columns
```

```python
df["diffrence"] = df[' Blood Pressure After (mmHg)'] - df[' Blood Pressure Before (mmHg)']
```

```python
sm.qqplot(df['diffrence'], line='s')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.title('Q-Q Plot')
plt.show()
```
